```
import matplotlib.pyplot as plt

# Data
age_groups = ['0-20 Years', '21-54 Years', '55+ Years']
population = [34, 57, 9]

# Create bar chart
plt.bar(age_groups, population)

# Labels and title
plt.xlabel('Age Group')
plt.ylabel('Population (%)')
plt.title('Population Distribution by Age')

# Show chart
plt.show()
```
